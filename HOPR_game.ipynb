```python
#Question 1: Function

def game():         #Here is the definition of the function called game. This function has no inputs here.
    Alice_tokens= 16    #The number of tokens with which Alice starts the game.
    Bob_tokens = 16     #The number of tokens with which Alice starts the game.
    tokens_tot = Alice_tokens + Bob_tokens   #The total tokens that Alice and Bob have.
    player_index = 1    # This index is created in order to help within the function to define which player's turn is.
    
    
    
    
    print('Alice starts the game!') #With this command the message in the parenthesis is printed.
    
    while tokens_tot<96:    # We use a while loop. Every loop resembles a round of the game running and every round will run if the total of tokens is below 96.
        
        if (player_index % 2) != 0:    # With this condition the code below is executed., If the rest of the division of the player_index is different than zero (meaning and odd number) Alice is playing the round.
            print('Alice plays in this round!')    #With this command the message in the parenthesis is printed.
            action = input('What action Alices chooses? Select "C" for continue or "S" for stop ')    # a string is stored in action variable using the input function. It is assigned in a textbox after the question appears on the screen.
                        
            if action == 'C':    #if 'C' was assigned to action the code below is executed.
                to_Bob = Alice_tokens/2    #Tokens that go to Bob, which are equal to what Alice has divided by two.
                Bob_tokens = Bob_tokens + to_Bob    #Total tokens of Bob at the end of this round.
                Alice_tokens = (Alice_tokens/2) + 16    #Total tokens of Alice at the end of this round, which equals to the half of that Alice had plus 16. 
                tokens_tot = Alice_tokens + Bob_tokens #Total tokens of Alice and Bob.
                player_index += 1 #After the end of this round 1 is added to player index in order to start the next round for the other player.
                               
            elif action == 'S': #if 'S' was assigned to action the code below is executed.
                
                print('Game over') #'Game over' is printed
                break #The while loop stops executing. The function ends and goes to return statement
                
            else: # If nor 'C' nor 'S'  was selected in the action, the code below is executed.
                print('Invalid argument. Please select only "C" for continue or "S" for stop')    #With this command the message in the parenthesis is printed.  
                
        else: # This else statement runs if the player_index is an even number. The code this else conditional statement is similar to the previous and it runs when it Bob is playing in the round.
            print('Bob plays in this round!')
            action = input('What action Bob chooses? Select "C" for continue or "S" for stop ')
                       
            if action == 'C':
                to_Alice = Bob_tokens/2
                Alice_tokens = Alice_tokens + to_Alice
                Bob_tokens = (Bob_tokens/2) + 16
                tokens_tot = Alice_tokens + Bob_tokens
                player_index += 1 
                               
            elif action == 'S':
                           
                print('Game over')
                break
            else:
                print('Invalid argument. Please select only "C" for continue or "S" for stop')  
# The return statement below is the output of the function. It is a message mentioning the amount of tokens that Alice and Bob own. Additionally it prints the total amount of tokens.
    return(print('Alice receives: ' + str(Alice_tokens)+ ' HOPR\n' + 'Bob receives: '+ str(Bob_tokens)+ ' HOPR' + '\nTotal amount of tokens in the game: ' + str(tokens_tot) + ' HOPR'))             
   
```

```python
# Calling the function to try it.

game()
```

```python
#Question 2

#In order to give all possible token allocations, it is important to define:

#1) Who was the player on the round
#2) Which round it was
#3) What was his action in this round
#4) Token allocation-How many tokens each player has at the end of the specific round and the total tokens in the round.

#The above information will be assigned in different lists
#There are two possible action in each round (C or S), that's why there are two possible token allocations for each round.

player = ['Alice', 'Alice', 'Bob', 'Bob', 'Alice', 'Alice', 'Bob', 'Bob', ]  #The player of the round
rounds = [1, 1, 2, 2, 3, 3, 4, 4] #Which round it could be
action = ['C', 'S', 'C', 'S', 'C', 'S', 'C', 'S'] #The possible action of the player
alice_tokens = [24, 16, 36, 24, 34, 36, 57, 34] #The possible token allocation for Alice at the end of the round
bob_tokens = [24, 16, 28, 24, 46, 28, 39, 46] #The possible token allocation for Bob at the end of the round
tot_tokens = [48, 32, 64, 48, 80, 64, 96 , 80] #The total token allocation

import pandas as pd #Importing pandas library in order to create a dataframe by the lists above.


# A dataframe is assigned in the variable df. With the command list(zip()) it creates a list of the above lists. Next name of the columns are given in the same order as the lists.

df = pd.DataFrame(list(zip(rounds, player, action, alice_tokens, bob_tokens, tot_tokens)), 
                  columns = ['Rounds', 'Player_in_this_round' , 'Action', 'Alice_tokens', 'Bob_tokens', 'Total_tokens'])
df.to_csv('HOPR_game.csv', index = False) # The dataframe is assigned and saved on a .csv file without indexing the rows. The path of the .csv will be the same that this.ipynb file is stored.
```

```python
#Question 3

# Tokens are collected by the players either when the total tokens is larger or equal to 96 or when one of the selects to stop.
# For this reason the dataframe in the previous question which contains all possible token allocations will be filtered.


df[(df.Total_tokens >= 96) | (df.Action == 'S')] # Filtering the dataframe in order to give numbers equal to 96 or larger from total tokens columns or all the 'S' actions from Action column.

# What can be observed is:
# 1) Alice will get the highest amount of tokens by Bob's action to continue on the 4th round. It is when the game ends automatically with a total of 96 tokens. Alice earns 57 tokens.
# 2) Alice will get the second highest amount, by her own action to stop on the 3rd round. Alice earns 36 tokens.
# 3) Bob will get the highest amount by choosing to stop on the 4th round. Bob earns 46 tokens.

# We can obtain the above values by coding. This will be done in the line below.
```

```python
# 1) Alice will get the highest amount of tokens if Bob would choose to continue on the 4th round

max_Alice = df['Alice_tokens'].max()
print(max_Alice)
print('')

# 2) Alice will get the second highest amount, by her own action to stop on the 3rd round. 
# In this case we can get the 2 largest numbers in the column of Alice_tokens and we already know know that the largest of all is 96. The second largest is the one we are looking for.
# The code below gives the two largest values in descending order. In addition gives the lines of the dataframe and the type of the objects in the selected column.

Alice_two = df['Alice_tokens'].nlargest(2)
print(Alice_two)
print('')

# 3) Bob will get the highest amount by choosing to stop on the 4th round.

max_Bob = df['Bob_tokens'].max()
print(max_Bob)
```
